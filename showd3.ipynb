# Use IPython & magic variables

IPython module has a display module which has
- function: display 
- classes: HTML, IFrame, Image, Javascript etc.

```python
from IPython.display import display                 # Import display function
from IPython.display import HTML, Image, Javascript     # Import classes
```

## Use Javascript

#### Works

```python
Javascript("element.text('1.hello world')")

Javascript("""
var i = 1
element.text(i)
""")
```

The output from the latest statement only gets displayed. But if we use **display** the prior statements also get displayed

```python
display(Javascript("element.text('hello world')"))
Javascript("""
var i = 1
element.text(i)
""")
```

```python
Javascript("console.log('hello world')")
```
